### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
import sympy as sp

# –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó
x = sp.Symbol('x') #—Å—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å—ñ–º–≤–æ–ª—å–Ω–æ—ó –∑–º—ñ–Ω–Ω–æ—ó
f = 10 * x - 5
df = sp.diff(f, x) #–ø–æ—Ö—ñ–¥–Ω–∞

print("–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó f(x) = 10x - 5 –¥–æ—Ä—ñ–≤–Ω—é—î:", df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)  # 100 —Ç–æ—á–æ–∫ –º—ñ–∂ 0 —ñ 10

f = 10 * x - 5               # –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó
df = np.full_like(x, 10)

plt.figure(figsize=(10, 5))

plt.plot(x, f, label='f(x) = 10x - 5', color='blue')        # —Ñ—É–Ω–∫—Ü—ñ—è
plt.plot(x, df, label="f'(x) = 10", color='green', linestyle='--')  # –ø–æ—Ö—ñ–¥–Ω–∞

plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó f(x) —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó f'(x)")
plt.xlabel("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è (x)")
plt.ylabel("–û—Ü—ñ–Ω–∫–∞")
plt.legend()
plt.grid(True)
plt.show()
```

–Ω–∞ –≥—Ä–∞—Ñ—ñ–∫—É –º–∏ –±–∞—á–∏–º–æ, —â–æ –∫–æ–∂–Ω–∞ –¥–æ–¥–∞—Ç–∫–æ–≤–∞ –≥–æ–¥–∏–Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è –∑–±—ñ–ª—å—à—É—î –Ω–∞—à—É –æ—Ü—ñ–Ω–∫—É. –î–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó f(x) –ø–æ—Ö—ñ–¥–Ω–∞ –¥–æ—Ä—ñ–≤–Ω—é—î 10.

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



```python
x = sp.Symbol('x')
funcion_sin = sp.sin(x)
funcion_ln = sp.ln(x)

#–ø–æ—Ö—ñ–¥–Ωi
df_sin = sp.diff(funcion_sin, x)
df_ln = sp.diff(funcion_ln, x)

print("–ü–æ—Ö—ñ–¥–Ω–∞ sin(x):", df_sin)
print("–ü–æ—Ö—ñ–¥–Ω–∞ ln(x):", df_ln)
```


3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
x = sp.Symbol('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x) #–ø–æ—Ö—ñ–¥–Ω–∞

print("–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó f(ùë•)=ùë•2‚àí5ùë•+1 –¥–æ—Ä—ñ–≤–Ω—é—î:", df)
```

```python
x = np.linspace(0, 100, 200)  # 100 —Ç–æ—á–æ–∫ –º—ñ–∂ 0 —ñ 10

f = x**2 - 5*x + 1              # –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó
df = np.full_like(x, 100)

plt.figure(figsize=(10, 5))

plt.plot(x, f, label='f(x) = x**2 - 5*x + 1', color='blue')        # —Ñ—É–Ω–∫—Ü—ñ—è
plt.plot(x, df, label="f'(x) = 2*x - 5", color='green', linestyle='--')  # –ø–æ—Ö—ñ–¥–Ω–∞

plt.title("–ì—Ä–∞—Ñ—ñ–∫ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó (–∑—Ä—ñ—Å—Ç/–ø–∞–¥—ñ–Ω–Ω—è –∑–∞–≥–∞–ª—å–Ω–∏—Ö –≤–∏—Ç—Ä–∞—Ç –∫–æ–º–ø–∞–Ω—ñ—ó) ")
plt.xlabel("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –º–∞—Ç–µ—Ä—ñ–∞–ª—ñ–≤ (x)")
plt.ylabel("–û—Ü—ñ–Ω–∫–∞")
plt.legend()
plt.grid(True)
plt.show()
```

–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó f(ùë•)=ùë•2‚àí5ùë•+1 –¥–æ—Ä—ñ–≤–Ω—é—î: 2*x - 5. –ó —Ü—å–æ–≥–æ –º–∏ –º–æ–∂–µ–º–æ –ø–æ—Ä–∞—Ö—É–≤–∞—Ç–∏, —â–æ –ø–æ—Ö—ñ–¥–Ω–∞ —Å—Ç–∞–Ω–æ–≤–∏—Ç—å -2.5 , —Ç–æ–±—Ç–æ x > -2.5 —â–æ –æ–∑–Ω–∞—á–∞—î - –∑—Ä—ñ—Å—Ç –≤–∏—Ç—Ä–∞—Ç –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞.
