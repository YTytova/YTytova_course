````python
import os, json
from langchain_openai import ChatOpenAI
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.agents import initialize_agent, AgentType
````

````python
search_tool = DuckDuckGoSearchRun()

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)
````

````python
agent = initialize_agent(
    tools=[search_tool],
    llm=llm,
    agent_type=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
````

````python
query = """
–ó–Ω–∞–π–¥–∏ 5 –æ—Å—Ç–∞–Ω–Ω—ñ—Ö –Ω–∞—É–∫–æ–≤–∏—Ö –ø—É–±–ª—ñ–∫–∞—Ü—ñ–π –Ω–∞ —Ç–µ–º—É "—à—Ç—É—á–Ω–∏–π —ñ–Ω—Ç–µ–ª–µ–∫—Ç".
–î–ª—è –∫–æ–∂–Ω–æ—ó –ø—É–±–ª—ñ–∫–∞—Ü—ñ—ó –≤–∫–∞–∂–∏:
1) –ù–∞–∑–≤—É —Å—Ç–∞—Ç—Ç—ñ
2) –ê–≤—Ç–æ—Ä—ñ–≤
3) –ö–æ—Ä–æ—Ç–∫–∏–π –æ–ø–∏—Å –∞–±–æ –≤–∏—Å–Ω–æ–≤–æ–∫.
–Ø–∫—â–æ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó –º–∞–ª–æ ‚Äî –¥–∞–π –Ω–∞–π–±–ª–∏–∂—á—ñ —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏.
"""
````

````python
# –ú–æ–¥–µ–ª—å
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)

# –ü–æ—à—É–∫ –∞–∫—Ç—É–∞–ª—å–Ω–æ—ó —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó (–ø–æ–≥–æ–¥–∞, –µ–∫–æ–Ω–æ–º—ñ–∫–∞)
search_tool = DuckDuckGoSearchRun()

# –Ü–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç Python –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—ñ–≤
def run_python(code: str):
    local_vars = {}
    try:
        exec(code, globals(), local_vars)
        return local_vars.get("result", "Python code –≤–∏–∫–æ–Ω–∞–Ω–æ")
    except Exception as e:
        return f"–ü–æ–º–∏–ª–∫–∞ —É –≤–∏–∫–æ–Ω–∞–Ω–Ω—ñ –∫–æ–¥—É: {e}"

python_tool = Tool(
    name="python_runner",
    func=run_python,
    description="–í–∏–∫–æ–Ω—É—î Python –∫–æ–¥ –¥–ª—è –±—ñ–∑–Ω–µ—Å-–∞–Ω–∞–ª—ñ—Ç–∏–∫–∏ —Ç–∞ –ø—Ä–æ–≥–Ω–æ–∑—ñ–≤."
)
````
